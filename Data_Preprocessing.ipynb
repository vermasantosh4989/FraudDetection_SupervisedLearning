# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Load Dataset

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head(5)
```

```python
df.tail()
```

```python
df.shape
```

```python
df.columns
```

```python
print(df["Class"].value_counts(normalize=True))
```

## Stratified Sample (50000 rows)

```python
sample_df,_ = train_test_split(
    df,
    train_size=50000,
    stratify=df["Class"],
    random_state=42
)
print(sample_df.shape)
```

```python
print(sample_df["Class"].value_counts(normalize=True))
```

```python
sample_df.head()
```

```python
sample_df.describe()
```

## Countplot

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=sample_df)
plt.xlabel("class")
plt.yscale("log")
plt.ylabel("class(log scale)")
plt.show()
```

## Amount Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(sample_df[sample_df.Class==0]["Amount"],color="blue",label="legitimate",bins=50)
sns.histplot(sample_df[sample_df.Class==1]["Amount"],color="red",label="Fraud",bins=50)
plt.legend()
plt.show()
```

## Time Distribution

```python
plt.figure(figsize=(6,5))
sns.histplot(sample_df[sample_df.Class==0]["Time"],color="blue",bins=50)
sns.histplot(sample_df[sample_df.Class==1]["Time"],color="blue",bins=50)
plt.legend()
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))
corr = sample_df[["V1","V2","V3","V4","V5","V6","V7","V8","V9","V10"]].corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

## Feature Engineering

```python
sample_df["Amount_log"]=np.log1p(sample_df["Amount"])
```

```python
sample_df["Amount"]
```

```python
sample_df["Amount_log"]
```

```python
sample_df["Hour"]=(sample_df["Time"]%86400)//3600
```

```python
sample_df["Hour"]
```

## Drop Columns

```python
sample_df.drop(["Amount","Time"],axis=1,inplace=True)
```

```python
sample_df.head()
```

## Scaling Dataset

```python
scaler=StandardScaler()
```

```python
sample_df[["Amount_log","Hour"]]=scaler.fit_transform(sample_df[["Amount_log","Hour"]])
```

```python
sample_df.head()
```

## Save Clean Data

```python
sample_df.to_csv("clean_data.csv",index=False)
```
